```python
import numpy
import matplotlib.pyplot as plt

#array maluco
n = 256
d = numpy.random.random(n)
d[128:140] = 2

#funcao vampirizada do stackoverflow
def reject_outliers(data, m=2, preencher=-1):
    data[abs(data - numpy.mean(data)) > m * numpy.std(data)] = preencher
    return data

#limpando o array maluco

n = reject_outliers(d, m=2, preencher=-1)

plt.plot(d,'r')
plt.plot(n,'k')
```
